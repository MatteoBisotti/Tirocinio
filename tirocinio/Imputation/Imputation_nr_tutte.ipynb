```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Separiamo le feature binarie dalle altre
binary_features = ['HIPRL', 'GENDER_01', 'Taglia', 'LUX_01', 'LUX_CR']
continuous_features = [col for col in dataset.columns if col not in binary_features]

# Preprocessamento: Normalizziamo solo le feature continue
scaler = StandardScaler()
dataset_scaled = dataset.copy()
dataset_scaled[continuous_features] = scaler.fit_transform(dataset[continuous_features].fillna(dataset[continuous_features].mean()))

# Prepariamo i dati per l'addestramento del modello
X_train = dataset_scaled.copy()

# Costruzione del modello di rete neurale
def build_model(input_dim):
    model = Sequential([
        Dense(10, activation='relu', input_dim=input_dim),
        Dropout(0.2),
        Dense(5, activation='relu'),
        Dropout(0.2),
        Dense(input_dim, activation='linear')  # Ricostruiamo tutti i valori
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    return model

# Addestramento del modello
model = build_model(X_train.shape[1])
model.fit(X_train, X_train, epochs=50, validation_split=0.2, verbose=0)

# Funzione per l'imputazione dei valori mancanti
def impute_missing_values(model, data, scaler, binary_features):
    continuous_features = [col for col in data.columns if col not in binary_features]
    mask = np.isnan(data.values)
    data_scaled = data.copy()
    data_scaled[continuous_features] = scaler.transform(data[continuous_features].fillna(data[continuous_features].mean()))
    
    data_imputed_scaled = model.predict(data_scaled)
    data_imputed = data.copy()
    data_imputed[continuous_features] = scaler.inverse_transform(data_imputed_scaled[:, :len(continuous_features)])
    
    for i, feature in enumerate(binary_features):
        data_imputed[feature] = np.where(data_imputed_scaled[:, len(continuous_features) + i] > 0.5, 1, 0)
    
    data_filled = data.copy()
    data_filled[mask] = data_imputed[mask]
    
    return data_filled

# Imputiamo i valori mancanti nel DataFrame originale
dataset_imputed = impute_missing_values(model, dataset, scaler, binary_features)
print(dataset_imputed.head())
```
