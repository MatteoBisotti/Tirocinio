```python
dataset = imp.clean_dataset(dataset)

dataset['DIRECTION'] = dataset['DIRECTION'].replace('CRANIO-DORSALE', 0)
dataset['DIRECTION'] = dataset['DIRECTION'].replace('CAUDO-VENTRALE', 1)

cols = ['n_luxation', 'first_lux_days_after_thr', 'DIRECTION']
dataset[cols] = dataset[cols].fillna(-1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
dataset = pd.DataFrame(scaler.fit_transform(dataset), columns=dataset.columns)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam


# Creazione del modello di rete neurale
def build_model(input_dim):
    model = Sequential([
        Dense(64, activation='relu', input_dim=input_dim),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='linear')  # Predice una singola feature
    ])
    model.compile(loss='mse', optimizer='adam')
    return model
```

```python
def impute_feature(df, feature):
    # Crea una copia del DataFrame
    df_copy = df.copy()
    
    # Divide i dati in feature e target
    X = df_copy.drop(columns=[feature]) # contiene tutte le colonne del DataFrame tranne la feature da imputare.
    y = df_copy[feature]                # contiene solo la colonna della feature da imputare

    # Identifica i valori mancanti nella feature da imputare
    missing_bool = y.isna()             # contiene valori booleani, true se il dato è mancante e false altrimenti
    
    # Separa i dati con e senza valori mancanti

    X_train = X[~missing_bool]          # contiene le righe di X dove i valori non sono mancanti
    y_train = y[~missing_bool]          # contiene le righe di y dove i valori non sono mancanti
    X_missing = X[missing_bool]         # contiene le righe di X dove i valori in y sono mancanti 
    
    # Costruzione e addestramento del modello
    model = build_model(X_train.shape[1])
    model.fit(X_train, y_train, epochs=100, batch_size=8, validation_split=0.2, verbose=0)
    
    # Predizione dei valori mancanti
    y_pred = model.predict(X_missing)
    
    # Imputazione dei valori mancanti
    y.loc[missing_bool] = y_pred.flatten()
    
    return y
```

```python
dataset['ALO'] =impute_feature(dataset, 'ALO')
```
