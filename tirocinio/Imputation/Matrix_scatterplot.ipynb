```python
import pandas as pd
import imputation as imp
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('/home/biso/tirocinio/Imputation/datiLussazioniDefinitivi.csv', delimiter=';')
dataset.head()
```

```python
dataset = imp.sub_dataset_cani(dataset)
dataset = imp.clean_dataset(dataset)
dataset = imp.drop_cols(dataset)
```

```python
dataset.head()
```

```python
dataset = dataset.drop(['INDICATIONFORTHR'], axis=1)
```

```python
corr_matrix = dataset.corr()

# Dimensione della figura
plt.figure(figsize=(28, 20))

# Informazioni della heatmap
sns.heatmap(corr_matrix, annot=True, 
                      fmt='.1f', cmap='coolwarm', 
                      square=True, linewidths=.5, 
                      cbar_kws={'shrink': .5},
                      vmin=-1.0, vmax=1.0)

# Titolo
plt.title("Matrice di correlazione")

# Descrizione delle etichette della matrice
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
```

```python
# matrice scatter plot
sns.pairplot(dataset)
plt.show()
```
