Il numero maggiore di Nan value lo troviamo nelle feature:
+ n_luxation
+ first_lux_days_after_thr
+ DIRECTION

I valori nan sono in corrispondenza di tutti quei cani che non hanno subito lussazione, è normale che il numero di lussazioni sia mancante per un cane che non ha subito lussazione dopo l'operazione.

Troviamo poi anche altri pochi valori Nan sparsi in qualche feature, che rappresentano dei valori mancanti.


```python
import missingno as msno

msno.matrix(dataset)
```

```python
import matplotlib.pyplot as plt

msno.bar(dataset)

plt.title("Visualizzazione del numero ndi Nan value presenti nel dataset", fontsize=30)

plt.show()
```

```python
dataset['LUX_01'].value_counts()
```

```python
import seaborn as sns

value_counts = dataset['LUX_01'].value_counts()

value_counts_dataset = value_counts.reset_index()
value_counts_dataset.columns = ['Value', 'Count']

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Value', y='Count', data=value_counts_dataset, palette='deep')

plt.xlabel('Classe')
plt.ylabel('Conteggio valori')
plt.show()
```
