### Confronto tra le matrici di correlazione 

```python
import pandas as pd

import sys 
sys.path.append("../base_lib")
import functions as func
```

```python
dataset = pd.read_csv("../csv/dataset_original.csv")
dataset_SMOTENC = pd.read_csv("../csv/dataset_SMOTENC.csv")
dataset_encoder = pd.read_csv("../csv/dataset_encoder.csv")

dataset = func.drop_cols(dataset)
dataset_SMOTENC = func.drop_cols(dataset_SMOTENC)
dataset_encoder = func.drop_cols(dataset_encoder)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 12))
plt.scatter(dataset['YEAR'], dataset['GENERATION'])
plt.xlabel('Anno')
plt.ylabel('Generazione della protesi')
plt.show()
```

```python
dataset['YEAR'].corr(dataset['GENERATION'])
```

```python
plt.figure(figsize=(15, 12))
plt.title("Scatter plot tra il peso del cane e la dimensione della protesi CUP")
plt.scatter(dataset['BODYWEIGHTKG'], dataset['CUPSIZE'])
plt.xlabel('Peso')
plt.ylabel('Dimensione della protesi')
plt.show()

print(dataset['BODYWEIGHTKG'].corr(dataset['CUPSIZE']))
```

```python
plt.figure(figsize=(15, 12))
plt.title("Scatter plot tra il peso del cane e la dimensione della protesi HEAD")
plt.scatter(dataset['BODYWEIGHTKG'], dataset['HEADSIZE'])
plt.xlabel('Peso')
plt.ylabel('Dimensione della protesi')
plt.show()

print(dataset['BODYWEIGHTKG'].corr(dataset['HEADSIZE']))
```

```python
plt.figure(figsize=(15, 12))
plt.title("Scatter plot tra il peso del cane e la dimensione della protesi NECK")
plt.scatter(dataset['BODYWEIGHTKG'], dataset['NECKSIZE'])
plt.xlabel('Peso')
plt.ylabel('Dimensione della protesi')
plt.show()

print(dataset['BODYWEIGHTKG'].corr(dataset['NECKSIZE']))
```

```python
import matplotlib.pyplot as plt

# Creazione della figura e dei subplot
fig, axs = plt.subplots(1, 3, figsize=(20, 6))  # 1 riga, 3 colonne

# Primo scatter plot
axs[0].scatter(dataset['BODYWEIGHTKG'], dataset['NECKSIZE'])
axs[0].set_title("Protesi NECK")
axs[0].set_xlabel('Peso')
axs[0].set_ylabel('Dimensione della protesi')

# Secondo scatter plot
axs[1].scatter(dataset['BODYWEIGHTKG'], dataset['HEADSIZE'])
axs[1].set_title("Protesi HEAD")
axs[1].set_xlabel('Peso')
axs[1].set_ylabel('Dimensione della protesi')

# Terzo scatter plot
axs[2].scatter(dataset['BODYWEIGHTKG'], dataset['CUPSIZE'])
axs[2].set_title("Protesi CUP")
axs[2].set_xlabel('Peso')
axs[2].set_ylabel('Dimensione della protesi')

# Regola lo spazio tra i plot
plt.tight_layout()

# Visualizza la figura
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creazione della figura e dei subplot
fig, axs = plt.subplots(2, 2, figsize=(18, 12))  # 2 righe, 2 colonne

# Primo scatter plot (prima riga, prima colonna)
axs[0, 0].scatter(dataset['BODYWEIGHTKG'], dataset['NECKSIZE'])
axs[0, 0].set_title("Scatter plot tra il peso del cane e la dimensione della protesi NECK")
axs[0, 0].set_xlabel('Peso  ')
axs[0, 0].set_ylabel('Dimensione della protesi')

# Secondo scatter plot (prima riga, seconda colonna)
axs[0, 1].scatter(dataset['BODYWEIGHTKG'], dataset['HEADSIZE'])
axs[0, 1].set_title("Scatter plot tra il peso del cane e la dimensione della protesi HEAD")
axs[0, 1].set_xlabel('Peso')
axs[0, 1].set_ylabel('Dimensione della protesi')

# Terzo scatter plot (seconda riga, seconda colonna)
axs[1, 0].scatter(dataset['BODYWEIGHTKG'], dataset['CUPSIZE'])
axs[1, 0].set_title("Scatter plot tra il peso del cane e la dimensione della protesi CUP")
axs[1, 0].set_xlabel('Peso')
axs[1, 0].set_ylabel('Dimensione della protesi')

# Nascondere l'asse vuoto (seconda riga, prima colonna)
axs[1, 1].axis('off')

# Regola lo spazio tra i plot
plt.tight_layout()

# Visualizza la figura
plt.show()
```
